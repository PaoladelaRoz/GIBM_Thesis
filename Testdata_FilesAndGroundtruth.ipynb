```python
import os

filepath = "data/data/test/lab/"
replace_map = {
    '1': 'Beatone',
    '2': 'Beattwo',
    '3': 'Beatthree',
    '4': 'Beatfour',
    '5': 'Beatfive'
}

def replace_numbers(line):
    for key, value in replace_map.items():
        line = line.replace(key, value)
    return line


files = os.listdir(filepath)

for filename in files:
    with open(filepath + filename, 'r') as file:
        lines = file.readlines()

    with open(filepath + filename, 'w') as file:
        for line in lines:
            modified_line = replace_numbers(line)
            file.write(modified_line)

    print(f"The following file was processed and modified {filename}")

print("All files were processed and modified correctly.")
```

```python
#CREATION OF FILES .LAB
import os

filepath = "data/data/test/lab/"
files = os.listdir(filepath)

for file in files:
    with open(filepath + file, 'r') as input_file:
        linestocopy = input_file.readlines()

    with open(filepath + file + '.lab', 'w+') as output_file:
        output_file.write('"*/' + file + '.lab"\n')
        for line in linestocopy:
            output_file.write(line)
        output_file.write(".\n")

    print(f"Se ha procesado el archivo {file} correctamente.")

print("Todos los archivos han sido procesados y modificados correctamente.")
```
