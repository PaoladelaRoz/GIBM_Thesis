```python
#lab files were already obtained in a python code before. -> no need to edit the groundtruth

#File processing of the output of HVite
#converting it into Processed

import os

file_path = "recout12.txt"

try:
    with open(file_path, 'r') as file:
        lines = file.readlines()
        new_lines = []
        for line in lines:
            if line.startswith('#!MLF!#') or line.startswith('"*/'):
                new_lines.append(line)
            else:
                parts = line.split()
                if len(parts) == 1 and parts[0] == '.':
                    new_lines.append(line)
                elif len(parts) == 4:
                    start = int(parts[0]) // 20000
                    end = int(parts[1]) // 20000
                    word = parts[2]
                    score = parts[3]

                    repeat_count = end - start

                    for _ in range(repeat_count):
                        new_lines.append(f'{word} {score}\n')

    output_file_path = "./Processed/recout12.txt"

    os.makedirs(os.path.dirname(output_file_path), exist_ok=True)
    with open(output_file_path, 'w') as output_file:
        output_file.writelines(new_lines)

except FileNotFoundError:
    print(f'File {file_path} not found')
except Exception as e:
    print(f'Error processing {file_path}: {e}')
```

```python
#Processed2 version 
import os

file_path = "./Processed/recout12.txt"
output_filepath = "./Processed2/recout12.txt"
os.makedirs(os.path.dirname(output_filepath), exist_ok=True)

try:
    with open(file_path, 'r') as file:
        lines = file.readlines()
    
    with open(output_filepath, 'w') as output_file:
        # Write the MLF header only once
        output_file.write('#!MLF!#\n')
        
        for line in lines:
            # Skip the MLF header
            if not line.startswith('#!MLF!#'):
                output_file.write(line)
                    
except FileNotFoundError:
    print(f'File {file_path} not found')
except Exception as e:
    print(f'Error processing {file_path}: {e}')
```

```python
#Processed 3 version
import os

file_path2 = "./Processed2/recout12.txt"
processed_dir = "./Processed3/"
processed_filepath = os.path.join(processed_dir, "recout12.txt")

# Crear el directorio si no existe
os.makedirs(processed_dir, exist_ok=True)

try:
    #file processing:
    with open(file_path2, 'r') as file:
        lines = file.readlines()
        if lines[0].strip() == '#!MLF!#':
            lines = lines[1:]
        new_lines = []
        for line in lines:
            line = line.strip()
            if line.startswith('"*/'):
                # save just the number
                number = line.split('/')[1].split('.')[0]
                new_lines.append(number)
            elif line:
                # save just the column with the beat type
                first_column = line.split()[0]
                new_lines.append(first_column)

    with open(processed_filepath, 'w') as file:
        for new_line in new_lines:
            file.write(new_line + '\n')

except FileNotFoundError:
    print(f'File {file_path2} not found')
except Exception as e:
    print(f'Error processing {file_path2}: {e}')
```

```python
#accuracy and confussion matrix

import os
import numpy as np
import sys

# Groundtruth files separated by .lab
filepath = "./lab"
files = os.listdir(filepath)


def processfiles(rec_filepath, output_filename):
    with open(rec_filepath, 'r') as rec_file:
        recout_lines = rec_file.readlines()

    
    # Variable initialization
    Beatone = 0
    Beatone_beattwo = 0
    Beatone_beatthree = 0
    Beatone_beatfour = 0
    Beatone_beatfive = 0
    Beattwo = 0
    Beattwo_beatone = 0
    Beattwo_beatthree = 0
    Beattwo_beatfour = 0
    Beattwo_beatfive = 0
    Beatthree = 0
    Beatthree_beatone = 0
    Beatthree_beattwo = 0
    Beatthree_beatfour = 0
    Beatthree_beatfive = 0
    Beatfour = 0
    Beatfour_beatone = 0
    Beatfour_beattwo = 0
    Beatfour_beatthree = 0
    Beatfour_beatfive = 0
    Beatfive = 0
    Beatfive_beatone = 0
    Beatfive_beattwo = 0
    Beatfive_beatthree = 0
    Beatfive_beatfour = 0

    i = 0
    while i < len(recout_lines):  # First line has the name of the .lab that should be compared to
        line = recout_lines[i].strip()  # Delete additional spaces
        if line.isdigit():  # If it contains a digit then it means it is the number of the file it's referring to
            current_file = line
            #print(current_file) - for development purpose
            gt_filepath = os.path.join(filepath, f"{current_file}")  # Get the file to compare
            #print(gt_filepath) - for development purpose

            if os.path.exists(gt_filepath):
                with open(gt_filepath, 'r') as gt_file:  # Read lines of groundtruth file
                    gt_lines = gt_file.readlines()

                    # Go over the recout file starting one line below (current line is the number of file)
                    j = i + 1  # To iterate recout file
                    aux = 0
                    while j < len(recout_lines) and aux < len(gt_lines):  # Till we get to at most the end of the file
                        line2 = recout_lines[j].strip()  # Get rid of additional spaces
                        if line2 == ".":  # If there is a '.' then it's the end of that file
                            break

                        gt_beat = gt_lines[aux].strip()
                        if line2 == 'Beatone':
                            if gt_beat == 'Beatone':
                                Beatone += 1
                            elif gt_beat == 'Beattwo':
                                Beatone_beattwo += 1
                            elif gt_beat == 'Beatthree':
                                Beatone_beatthree += 1
                            elif gt_beat == 'Beatfour':
                                Beatone_beatfour += 1
                            elif gt_beat == 'Beatfive':
                                Beatone_beatfive += 1
                        elif line2 == 'Beattwo':
                            if gt_beat == 'Beattwo':
                                Beattwo += 1
                            elif gt_beat == 'Beatone':
                                Beattwo_beatone += 1
                            elif gt_beat == 'Beatthree':
                                Beattwo_beatthree += 1
                            elif gt_beat == 'Beatfour':
                                Beattwo_beatfour += 1
                            elif gt_beat == 'Beatfive':
                                Beattwo_beatfive += 1
                        elif line2 == 'Beatthree':
                            if gt_beat == 'Beatthree':
                                Beatthree += 1
                            elif gt_beat == 'Beattwo':
                                Beatthree_beattwo += 1
                            elif gt_beat == 'Beatone':
                                Beatthree_beatone += 1
                            elif gt_beat == 'Beatfour':
                                Beatthree_beatfour += 1
                            elif gt_beat == 'Beatfive':
                                Beatthree_beatfive += 1
                        elif line2 == 'Beatfour':
                            if gt_beat == 'Beatfour':
                                Beatfour += 1
                            elif gt_beat == 'Beattwo':
                                Beatfour_beattwo += 1
                            elif gt_beat == 'Beatthree':
                                Beatfour_beatthree += 1
                            elif gt_beat == 'Beatone':
                                Beatfour_beatone += 1
                            elif gt_beat == 'Beatfive':
                                Beatfour_beatfive += 1
                        elif line2 == 'Beatfive':
                            if gt_beat == 'Beatfive':
                                Beatfive += 1
                            elif gt_beat == 'Beattwo':
                                Beatfive_beattwo += 1
                            elif gt_beat == 'Beatthree':
                                Beatfive_beatthree += 1
                            elif gt_beat == 'Beatfour':
                                Beatfive_beatfour += 1
                            elif gt_beat == 'Beatone':
                                Beatfive_beatone += 1
                        aux += 1
                        j += 1
            i = j + 1
        else:
            i += 1

    CorrectBeats = (Beatone + Beattwo + Beatthree + Beatfour + Beatfive)
    IncorrectBeatone = Beatone_beattwo + Beatone_beatthree + Beatone_beatfour + Beatone_beatfive
    IncorrectBeattwo = Beattwo_beatone + Beattwo_beatthree + Beattwo_beatfour + Beattwo_beatfive
    IncorrectBeatthree = Beatthree_beatone + Beatthree_beattwo + Beatthree_beatfour + Beatthree_beatfive
    IncorrectBeatfour = Beatfour_beatone + Beatfour_beattwo + Beatfour_beatthree + Beatfour_beatfive
    IncorrectBeatfive = Beatfive_beatone + Beatfive_beattwo + Beatfive_beatthree + Beatfive_beatfour

    Totalbeats = CorrectBeats + IncorrectBeatone + IncorrectBeattwo + IncorrectBeatthree + IncorrectBeatfour + IncorrectBeatfive
    Accuracy = CorrectBeats / Totalbeats if Totalbeats > 0 else 0
    percentageAccuracy = Accuracy * 100

    if not os.path.exists(output_filename):
        open(output_filename, 'w').close() 

    with open(output_filename, 'w') as output_file:
        original_stdout = sys.stdout  # Save original stdout
        sys.stdout = output_file
        print(f"Correct beats: {CorrectBeats}") 
        print(f"Total beats: {Totalbeats}")
        print(f"Accuracy: {Accuracy:.2f} ({percentageAccuracy:.2f}%)")
        
        sys.stdout = original_stdout  # Restore original stdout

    print(f"Output saved to {output_filename}")

rec_filepath = f"./Processed3/recout12.txt"
output_filename =f"./Results/10result12.txt"
processfiles(rec_filepath, output_filename)
```
