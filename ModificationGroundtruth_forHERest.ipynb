```python
def unify_file(input_file, output_file):
    with open(input_file, 'r') as file:
        lines = file.readlines()
    
   
    unique_lines = []
    previous_line = None
    for line in lines:
        if line != previous_line:
            unique_lines.append(line)
        previous_line = line
    
   
    with open(output_file, 'w') as file:
        file.writelines(unique_lines)


input_file = 'groundtruth1.txt'
output_file = 'newfiles/groundtruth1.txt'
unify_file(input_file, output_file)
```
